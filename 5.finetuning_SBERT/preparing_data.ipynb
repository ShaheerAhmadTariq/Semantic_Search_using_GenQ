```python
import pandas as pd

df = pd.read_csv('../3.Query_Generation/dataset_Q_D.csv')
print(df.shape)
df.head()
```

### Creating a dataframe with specific columns: id, document, and query. Each query is a new row, repeating the document for each.

```python

rows = []
for i in range(len(df)):
    document = df.loc[i, 'document']
    queries = [df.loc[i, f'query{j}'] for j in range(1, 6)]
    for query in queries:
        rows.append({'id': i, 'document': document, 'query': query})

# Create the DataFrame from the list of rows
new_df = pd.DataFrame(rows)

# Display the size and the first few rows of the new DataFrame
print(new_df.shape)
new_df.head(10)
```

```python
# drop all rows with empty document or query
new_df = new_df.dropna(subset=['document', 'query'])
print(new_df.shape)
```

```python
new_df.to_csv('./dataset_D_Q.csv', index=False)
```
